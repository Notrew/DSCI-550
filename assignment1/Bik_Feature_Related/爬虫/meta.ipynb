```python
from selenium import webdriver
from bs4 import BeautifulSoup

chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome('C:/Program Files (x86)/Google/Chrome/Application/chromedriver.exe', options=chrome_options)
driver.get('https://www.imdb.com/title/tt0111161/')
```

```python
html = driver.page_source #获取页面源代码
```

```python
soup = BeautifulSoup(html)
print(soup.prettify()) #soup解析
```

```python
title = soup.title.string #按tag获取值
title
```

```python
dp = soup.find('meta',{'name':'description'}) #按照tag+attributes获取值
dp['content'] #获取content属性
```
